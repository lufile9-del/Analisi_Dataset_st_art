E' emerso che abbiamo diversi valori nulli, soprattutto nelle colonne `movimenti`, `data_creazione`, `altezza`, `larghezza`, `generi` e `soggetti`.



Per il momento ho deciso di non eliminare tutti i valori nulli nella loro totalità, in quanto essi potrebbero riguardare colonne che non interessano la mia analisi.


###3.3 Convesione `data_creazione` in **`Int64`**

```python
df["data_creazione"].info()
```

Per prima cosa estraggo i numeri di quattro cifre (gli anni di realizzazione delle opere) da `data_creazione`.

```python
df["anno_creazione"]=df["data_creazione"].str.extract(r"(\d{4})")
```

In secondo luogo vado a convertire la nuova colonna `anno_creazione` in `Int64`.


```python
df['anno_creazione'] = df['anno_creazione'].astype(pd.Int64Dtype())
```

Andiamo ora a verificare che tutto sia stato fatto correttamente:

```python
df["anno_creazione"].info()
```

### 3.3 Pulizia della colonna `artisti`

La colonna `artisti` continene al suo interno due informazioni differenti: il nome dell'artista ed il genere (uomo/femmina). Per una lettura più pulita ed efficiente ho deciso di creare una nuova colonna contenente solo il genere dell'artista (`genere_artista`), eliminandolo dalla colonna iniziale.

Per prima cosa estraiamo dalla colonna artisti i termini *(maschio)* | *(femmina)*. Essi dovrebbero essere esplicitati tutti in lettere minuscole, ma per essere sicuri che per essere sicuri che venga estratto tutto usiamo `(?i)`.

```python
df['genere_artista'] = df['artisti'].str.extract(r'(?i)\((maschio|femmina)\)', expand=False)
```

Ora puliamo la colonna artisti, eliminando i termini (*maschio*) | (*femmina*) , così come gli eventuali spazi bianchi residui.

```python
df['artisti'] = df['artisti'].str.replace(r'(?i)\s*\((?:maschio|femmina)\)', '', regex=True).str.strip()
```

Visualizziamo ora che tutto sia stato fatto correttamente:

```python
df.head()
```

##4.2 Explanatory Analysis

### 4.2.1 Indagine comparata
Seguendo questo ragionamento e considerando che a partire dal 1700 la differenza tra i due generi artistici è maggiormente accentuata, vado ad analizzare e visualizzare in maniera comparata l'andamento dell'arte religiosa e di quella laica nel dettaglio.  

Creiamo un nuovo DataFrame `df_periodo_filtro` che racchiuda il nostro periodo di interesse.

```python
df_periodo_filtro = df2.loc[
    (df2['anno_creazione'].notna()) &
    (df2['anno_creazione'] >= 1700) &
    (df2['anno_creazione'] <= 2000)
]
```

Visualizziamo che tutto sia stato svolto correttamente e verifichiamo la presenza di valori nulli, per essere sicuri.

```python
df_periodo_filtro.sort_values(by="anno_creazione")
```

```python
df_periodo_filtro.isnull().value_counts()
```

Realizziamo ora un grafico a linee, che faccia vedere l'andamento dei due generi.

```python
comparative_data = df_periodo_filtro.groupby(['generi', 'anno_creazione']).size().unstack(level=0)

comparative_data = comparative_data.fillna(0)

ax = comparative_data.plot.line(
    figsize=(14, 7),
    title='Andamento dell\'Arte Religiosa e dell\'Arte Laica (1700-2000)',
    xlabel='Anno di Creazione',
    ylabel='Numero di Opere',
    grid=True,
    color={'arte religiosa': '#1E90FF', 'arte laica': '#8B008B'}
)

ax.legend(title='Tipo di Arte', loc='upper right')
```

Il grafico evidenzia la situazione già ipotizzata. L'arte religiosa ha un un considerevole picco nel 1700, per poi ridurre le proprie attestazioni nel corso dei secoli successivi, salvo qualche picco minore.

L'arte laica, invece, ha una produzione maggiore e costante, che vede i propri picchi più importanti tra la seconda metà del 1800 e l'inizio del 1900.

Anche in questo caso il grafico ci permette di restituire una situazione più puntuale ed in linea con l'evoluzione della Storia dell'arte. A partire dalla seconda metà del '800 correnti come l'Impressionismo (e tutte le sue declinazioni), la graduale la rottura con l'accademismo e la ricerca di nuovi valori, portano inevitabilmente gli artisti ad allontanarsi da temi canonici e tradizionali, in cui rientrano anche quelli sacri.

Vediamo ora il rapporto nello specifico tra arte laica ed arte religiosa in questo lasso di tempo tramite un grafico a torta.

```python
df2_periodo_filtro=df_periodo_filtro['generi'].value_counts()

ax= df2_periodo_filtro.plot.pie(
    title="Arte Religiosa e Arte Laica (1700-2000)",
    autopct='%1.1f%%',
    startangle=90,
    figsize=(6, 6),
    ylabel="",
    explode=(0.02,0.02),
    textprops={'fontsize': 12, 'color': 'black'},
    colors=  ['#8B008B', '#1E90FF']
)
```

Il grafico a torta restituisce in maniera ancora più lampante ciò che è stato detto ed ipotizzato sino a questo punto.

**La produzione artistica di arte sacra tra il 1700 e il 2000 è circa 1/5 di quella laica!**

####*Chissà se c'è una "differenza" anche della predilizione di un genere specifico dell'arte laica piuttosto che un altro?*

Anche in questo caso filtro la produzione maschile, eliminando tutte le opere che rientrano nell'arte religiosa.

```python
generi_completi_uomini = df_uomini[df_uomini["generi"] != "arte religiosa"]
generi_completi_uomini
```

Creo un nuovo dataframe, contenente gli artisti filtrati (`df_uomini_generi`) ed i generi "esplosi" (`df_generi`)

```python
df_uomini_generi= generi_completi_uomini[["artisti"]].join(df_generi)
```

Poichè anche in questo caso si creano "doppioni" dovuti allo sdoppiamento dei generi, ho deciso di eliminare il primo di ogni duplicato, trattandosi comunque di generi simili tra di loro, poichè a me interessa avere una visione generale sui diversi generi laici trattati dagli uomini, per fare un confronto con la produzione femminile.

```python
df_uomini_generi = df_uomini_generi[~df_uomini_generi.index.duplicated(keep='first')]
df_uomini_generi.head()
```

Realizzo ora un grafico a torta.
I generi realizzati dagli uomini sono maggiori e diversificati. Al fine di ottenere un grafico quando più pulito possibile ho deciso di racchiudere tutti i generi con un'attestazione minore al 5% sotto la dicitura "generi minori".

```python
conteggi_generi= df_uomini_generi['generi'].value_counts()

total_opere = conteggi_generi.sum()

soglia_minima_conteggio = total_opere * 0.05

generi_minori_conteggi = conteggi_generi[conteggi_generi < soglia_minima_conteggio]

generi_maggiori_conteggi = conteggi_generi[conteggi_generi >= soglia_minima_conteggio]

conteggi_finali = generi_maggiori_conteggi.copy()
if not generi_minori_conteggi.empty:
    conteggi_finali['generi minori'] = generi_minori_conteggi.sum()

ax = conteggi_finali.plot.pie(
    title='Artisti e Generi "laici"',
    autopct='%1.1f%%',
    startangle=90,
    figsize=(6, 6),
    ylabel='',
    explode=[0.02] * len(conteggi_finali),
    textprops={'fontsize': 10, 'color': 'black'},
    pctdistance=0.6,
    colors= ['#006400', '#228B22', '#6B8E23', '#9ACD32', '#90EE90']
)


ax.set_title(ax.get_title(), fontsize=12)
```

I generi realizzati dagli uomini appaiono molto più diversificati, come si ipotizzava. In particolare emerge la pittura mitologica, che nella pittura femminile non sembra apparire. Questo potrebbe essere legato a molteplici ipotesi: l'arte mitologica era spesso frutto di commissioni, per i valori impliciti che essa conteneva e veicolava, ed inoltre molto spesso conteneva soggetti nudi o semi-nudi.
Inoltre tra i generi minori della produzione maschile diverse sono le tipologie non presenti nei quadri delle pittrici. Primo fra tutti spicca il nudo artistico, con attestazioni pari al 3.8%.

Tali risultati si ricollegano perfettamente a ciò che è stato evidenziato nella produzione femminile.
